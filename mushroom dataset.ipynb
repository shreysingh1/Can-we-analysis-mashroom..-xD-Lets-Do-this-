# Encoding the values to make computer understand the values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Encoder_x=LabelEncoder()
for cols in X.columns:
    X[cols]=Encoder_x.fit_transform(X[cols])
Encoder_y=LabelEncoder()
Y=Encoder_y.fit_transform(Y)
```

# Lets Build the KNN model and separate the data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,Y)
```

# Lets find the most accurate value of K

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
acc = []
for neighbors in range(3,10):
    classifier = KNeighborsClassifier(n_neighbors=neighbors, metric='euclidean')
    classifier.fit(train_x,train_y)
    y_pred = classifier.predict(test_x)
    acc.append(accuracy_score(test_y,y_pred))
    
plt.figure(figsize=(15,7))
plt.plot(list(range(3,10,1)), acc)
plt.show()


```

```python
print(f"Best accuracy is {np.max(acc)} and the k value is {1+acc.index(np.max(acc))}")
```

```python
k=1+acc.index(np.max(acc))
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(train_x,train_y)
pred=knn.predict(test_x)
```
